```python
import pandas as pd

def add_century_column(df, birth_column='birth', new_column='century'):
    """
    Adds a column indicating the century based on the birth year, 
    considering both BC and AC dates.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the birth year column.
    birth_column (str): The name of the column with birth years.
    new_column (str): The name of the new column to store century information.

    Returns:
    pd.DataFrame: The DataFrame with the added century column.
    """
    def calculate_century(year):
        if pd.isnull(year):  # Handle missing data
            return None
        year = int(year)
        if year > 0:
            century = (year - 1) // 100 + 1
            return f"{century}th century AC"
        else:
            century = abs(year) // 100 + 1
            return f"{century}th century BC"

    # Apply the century calculation to the specified birth column
    df[new_column] = df[birth_column].apply(calculate_century)
    return df
```

```python
import pandas as pd

def add_century_columns(df, birth_column='birth'):
    """
    Adds two columns:
    - 'century': The century label (e.g., "20th century AC" or "5th century BC").
    - 'century_numeric': A numeric value for sorting (e.g., -5 for 5th century BC, 3 for 3rd century AC).

    Parameters:
    df (pd.DataFrame): The DataFrame containing the birth year column.
    birth_column (str): The name of the column with birth years.

    Returns:
    pd.DataFrame: The DataFrame with the added 'century' and 'century_numeric' columns.
    """
    def calculate_century(year):
        if pd.isnull(year):
            return None
        year = int(year)
        if year > 0:
            century = (year - 1) // 100 + 1
            return f"{century}th century AC"
        else:
            century = abs(year) // 100 + 1
            return f"{century}th century BC"

    def calculate_century_numeric(year):
        if pd.isnull(year):
            return None
        year = int(year)
        if year > 0:
            return (year - 1) // 100 + 1  # AC centuries as positive values
        else:
            return -((abs(year) // 100) + 1)  # BC centuries as negative values

    # Create the 'century' and 'century_numeric' columns in the DataFrame
    df['century'] = df[birth_column].apply(calculate_century)
    df['century_numeric'] = df[birth_column].apply(calculate_century_numeric)
    
    return df
```
